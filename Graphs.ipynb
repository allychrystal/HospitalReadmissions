```python
corrs = (df.corr()['readmitted'].apply(lambda x: abs(x)).sort_values(ascending=False).head(10))
```

```python
newdf = pd.DataFrame()
newdf['Correlations'] = corrs
newdf['Feature'] =corrs.index
newdf.drop('readmitted', inplace= True)
```

```python
plt.figure(figsize=(10, 5))
sns.barplot('Correlations', 'Feature', data= newdf, palette='Greens')
plt.title('Correlation Feature Importance', size= 15)
plt.xlabel('Correlation (Absolute Value)', size= 15)
plt.ylabel('')
sns.plt.xlim(0, 1)
plt.xticks(rotation=0, size=15)
```
